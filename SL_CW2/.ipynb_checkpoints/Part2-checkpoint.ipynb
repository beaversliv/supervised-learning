```python
import numpy as np
import pandas as pd
import time
from datetime import datetime
from scipy.spatial.distance import cdist
from numba import jit
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# Generate data
def generate_data(m,n,classifier):
    if classifier == 'winnow':
        data = np.random.randint(2, size=(m,n))
        data = np.sign(data)
    else:
        data = np.random.uniform(low=-1, high=1, size=(m,n))
        data = np.sign(data)

    x = data
    y = data[:,0]
    return x,y
```

## Binary Perceptron without kernel

```python
def perceptron_epoch(x, y, weight):
    """
    Given a set of weights alpha, this function goes through the whole set of examples in x, one by one
    and iteratively update the weights if a mistake has happened.
    :param x: the observations array
    :param y: the labels vector
    :param y_arr: the transformed labels matrix
    :param alpha: the set of weights
    :param K_train: the kernel matrix as calculated over x
    :return: the updated weights alpha and the number of errors made while going through the points
    """
    # Number of examples
    m = x.shape[0] 
    num_errors = 0

    for t in range(m):
        # Find our training set
        x_t = x[t,:] #n size (1,10)
        y_t = y[t]
        # pred_t computes \sum^{t-1}_{i=0} {(alpha_i K(x_t, x_i))}, 
        # which is regarded as the confidence in each class
        confidence = weight @x_t
        
        if confidence!=y_t:
            weight = weight+y_t*x_t
            num_errors+=1
    return weight, num_errors
```

```python
def perceptron_train(x, y,convergence_threshold=0.01):
    """
    Trains a perceptron based on the one vs. rest approach, i.e. train k classifiers to identify k classes
    :param x: the observations array
    :param y: the labels vector
    :param convergence_threshold: the threshold value upon we stop updating the perceptron, if the difference
    in errors was smaller than that
    :return:
    """
    m = x.shape[0] #number of examples
    n = x.shape[1] #number of features
    
    error_per_epoch = []
    weight = np.zeros((1,n)) #Need to store alpha array at all iteration, as we need it to compute confidence
    
       
    epochs = 0
    while True:
        weight, num_errors = perceptron_epoch(x,y,weight)
        
        error_rate_current = error_per_epoch[-1] / x.shape[0] if epochs > 0 else 0
        error_rate_next = num_errors / x.shape[0]

        error_per_epoch.append(num_errors)
        if epochs > 0 and (error_rate_next > error_rate_current or \
            error_rate_current - error_rate_next < convergence_threshold):
            break
            
        epochs += 1

    return weight, error_per_epoch
```

```python
def perceptron_test(x_test,y_test, weight):
    """
    This function predicts the labels for a given set of observations.
    :param x_test: the set of observations to predict
    :param x_train: the observations in which the perceptron has been trained on
    :param y_test: the true labels of x_test
    :param alphas: the set of weights of the perceptron
    :param d: is either the order of the polynomial kernel or the constant in the gaussian kernel
    :param kernel_choice: Depending on the kernel choice, can be {'Polynomial', 'Gaussian'}
    :return: the number of mistakes that the perceptron made, the predictions as well as the confidence values.
    """
    # Compute the confidence
    confidence = (weight @ x_test.T)
    
    # Compute mistakes and predictions
    preds = np.sign(confidence)
    diff = preds-y_test
    mistakes = len(np.flatnonzero(diff))
    return mistakes, preds
```

## winnow binary classification

```python
def winnow_epoch(x, y, weight):
    """
    Given a set of weights alpha, this function goes through the whole set of examples in x, one by one
    and iteratively update the weights if a mistake has happened.
    :param x: the observations array
    :param y: the labels vector
    :param y_arr: the transformed labels matrix
    :param alpha: the set of weights
    :param K_train: the kernel matrix as calculated over x
    :return: the updated weights alpha and the number of errors made while going through the points
    """
    # Number of examples
    m = x.shape[0]
    n = x.shape[1] 
    num_errors = 0

    for t in range(m):
        # Find our training set
        x_t = x[t,:] #n size (1,10)
        y_t = y[t]
        # pred_t computes \sum^{t-1}_{i=0} {(alpha_i K(x_t, x_i))}, 
        # which is regarded as the confidence in each class
        confidence = weight @x_t
        
        if confidence< n/2:
            pred_t = 0
        else:
            pred_t = 1
        if pred_t != y_t:
            power = (y_t-pred_t)*x_t/1.0
            weight = weight*(2**power)
            num_errors+=1
    # Store alpha_t into the matrix for future reference
    return weight, num_errors
```

```python
def winnow_train(x,y,convergence_threshold=0.01):
    """
    Trains a perceptron based on the one vs. rest approach, i.e. train k classifiers to identify k classes
    :param x: the observations array
    :param y: the labels vector
    :param convergence_threshold: the threshold value upon we stop updating the perceptron, if the difference
    in errors was smaller than that
    :return:
    """
    m = x.shape[0] #number of examples
    n = x.shape[1] #number of features
    
    error_per_epoch = []
    weight = np.ones((1,n)) #Need to store alpha array at all iteration, as we need it to compute confidence
    
       
    epochs = 0
    while True:
        weight, num_errors = winnow_epoch(x,y,weight)
        
        error_rate_current = error_per_epoch[-1] / x.shape[0] if epochs > 0 else 0
        error_rate_next = num_errors / x.shape[0]

        error_per_epoch.append(num_errors)
        if epochs > 0 and (error_rate_next > error_rate_current or \
            error_rate_current - error_rate_next < convergence_threshold):
            break
            
        epochs += 1

    return weight, error_per_epoch
```

```python
def winnow_test(x_test,y_test,weight):
    m = x_test.shape[0] #number of examples
    n = x_test.shape[1] #number of features
    confidence = weight @ x_test.T
    pred = np.zeros((1,m))
    pred[confidence>=n/2] = 1
    diff = pred - y_test
    mistakes = len(diff[diff!=0])
    return mistakes,pred
```

## least squares

```python
def least_squares(x,y,x_test,y_test):
    
    beta = np.linalg.pinv(x.T @ x) @ x.T @ y
    f = x_test @ beta
    preds = np.where(f>0,1,-1)
    diff = preds - y_test
    mistakes = len(np.flatnonzero(diff))
    return mistakes,preds
```

## 1-NN

```python
def kNN_test(x_train,y_train,x_test,y_test):
    p1 = np.sum(x_train**2,axis=1)[:,np.newaxis]
    p2 = np.sum(x_test**2,axis=1)
    p3 = -2*np.dot(x_train,x_test.T)
    Distances = p1+p2+p3
    
    pred = y_train[Distances.argmin(axis=0)]
    diff = pred - y_test
    mistakes = len(np.flatnonzero(diff))
    return mistakes,pred
```

## Estimate Sample Complexity

```python
def estimate_generalization_error(m,n,number_of_samples,classifier,test_size):
    """
    This function estimates the generalization error for fixed m and n, by sampling.
    :param m: int
    :param n: int
    :param number_of_samples: int, number of running time involved in estimation of generalizaiton error,usually set to 10
    :param functon_choice:  str, name of algorithm
    return: sclar for specific classifier
    
    """
    test_errors=np.zeros(number_of_samples)
    for run in range(number_of_samples):
        if classifier == 'winnow':
            x_test,y_test = generate_data(test_size,n,classifier)
        else:
            x_test,y_test = generate_data(test_size,n,classifier)
        # calculate one single error
        if classifier=='winnow':
            x,y = generate_data(m,n,classifier)
            W, _ = winnow_train(x,y,convergence_threshold=0.01)
            test_mistakes,_ = winnow_test(x_test,y_test,W)
        elif classifier == 'perceptron':
            x,y = generate_data(m,n,classifier)
            W, _ = perceptron_train(x, y,convergence_threshold=0.01)
            test_mistakes,_= perceptron_test(x_test,y_test,W)
        elif classifier == 'linear':
            x,y = generate_data(m,n,classifier)
            test_mistakes,_= least_squares(x,y,x_test,y_test)
        elif classifier == '1nn':
            x,y = generate_data(m,n,classifier)
            test_mistakes,_ = kNN_test(x,y,x_test, y_test)

        mean_test_error = (1.0/ y_test.shape[0]) * test_mistakes

        test_errors[run] = mean_test_error          
    gen_error = test_errors.mean()
        
    return gen_error
```

```python
def binary_search(m_left,m_right,max_n,classifier,error_threshold,step_size):
    number_of_runs=10
    sample_complexity = np.zeros((max_n,number_of_runs))
    
    over_estimates = 0
    under_estimates = 0
    exact_estimates = 0
    
    number_of_samples = 10
    test_size = 1000
    
    # define m range, may due to different classifier  
    for n in np.arange(1,max_n+1,step_size):
        for run in range(number_of_runs):
            
            m_low, m_high = m_left, m_right
            while True:
                m = (m_low + m_high) // 2
                
                gen_error = estimate_generalization_error(m,n,number_of_samples,classifier,test_size) 
                if gen_error > error_threshold:
                    if m_high - m < 2:
                        # binary search converged
                        under_estimates +=1                       
                        break
                    else:
                        # increase the lower search bound
                        m_low = m                         
                else: #error less than 10%
                    if m_high - m_low < 2: 
                        # binary search converged
                        if gen_error != 0.1:
                            over_estimates +=1
                        else:
                            exact_estimates +=1
                        break
                    else:
                        # decrease the upper search bound
                        m_high = m
                        
            sample_complexity[n-1,run] = m
                
    return sample_complexity
```

```python
def plot_sample_complexity(sample_complexity,step_size,classifier):
    """
    This function sample complexity estimated by Binary Search 
    :param: sample_complexity, 2D array that records the sample complexity computed for each n in each run
    :param step_size: int, step size to calculate sample complexity for n up to max_n
    :param function_choice: str, name of algorithm
    """
    # average along row
    means = np.mean(sample_complexity, axis=1)
    means_crop = means[means>0]
    errors = np.std(sample_complexity, axis=1)
    errors_crop = errors[means>0]
    
    plt.errorbar(np.arange(1,sample_complexity.shape[0]+1,step_size), means_crop, errors_crop, marker='^')
    plt.xlabel('n')
    plt.ylabel('sample complexity')
    plt.title('Estimated sample complexity for '+str(classifier))
```

```python
#linear
startTime = datetime.now()
linear_result = binary_search(m_left=1,m_right=700,max_n=100,classifier='linear',error_threshold=0.1,step_size=2)
plot_sample_complexity(linear_result,step_size=2,classifier='linear')
Linear = datetime.now() - startTime
print("\nTime taken: ", Linear)
```

```python
#1nn
startTime = datetime.now()
lnn_result = binary_search(m_left=1,m_right=25000,max_n=20,classifier='1nn',error_threshold=0.1,step_size=2)
plot_sample_complexity(lnn_result,step_size=2,classifier='1nn')
lnn = datetime.now() - startTime
print("\nTime taken: ", lnn)
```

## (C) Estimate m(n)

```python
def fitting_mn(sample_complexity,step_size,n_max,fitting_type,classifier):
    """
    This function is used to estimate m(n) where m is a function about n
    :param: sample_complexity 2D array with size(n_max,number_of_runs), calculated before
    :param:fitting_type,observed shape type.{'linear','exponential','log'}
    :param: classifier,{'linear','perceptron','winnow','1nn'}
    """
    x_range = np.arange(1,n_max,step_size)
    m_means = np.mean(sample_complexity,axis=1)
    # take elements that non-zero
    means_crop = m_means[m_means>0]
    
    if fitting_type=='linear':
        # fit with degree 1
        k,b = np.polyfit(x_range,means_crop,1)
        fit =  [b + k*x  for x in x_range]
        print('estimated (m(n))=',round(b,2),'+',round(k,2),'n')
    elif fitting_type=='exponential':
        k,b = np.polyfit(x_range,np.log(means_crop),1)
        fit =  [np.exp(b + k*x)  for x in x_range]
        print('estimated (m(n))=',round(np.exp(b),2),'exp(',round(k,2),'n)')
       
      
    elif fitting_type=='log':
        k,b = np.polyfit(np.log(x_range),means_crop,1)
        fit = [b + k*np.log(x)  for x in np.arange(1,100,step_size)]
        print('estimated (m(n))=',round(b,2),'+',round(k,2),'log(n)')
        
    plt.plot(x_range,fit,color='r',label='fitted')
    plt.plot(x_range,means_crop,color='b',label='actual')
    plt.title('Sample Complexity with ' + fitting_type +' fitting for '+ classifier)
    plt.ylabel('sample complexity')
    plt.xlabel('n')
    plt.show()
```

```python
fitting_mn(sample_complexity=linear_result, step_size=2,n_max=100,fitting_type='linear',classifier='linear')
```
